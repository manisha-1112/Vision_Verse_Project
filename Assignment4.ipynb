```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread(r'/content/lowContrast.png', cv2.IMREAD_GRAYSCALE)
laplacian_filtered = cv2.Laplacian(image, cv2.CV_64F)
laplacian_filtered = cv2.convertScaleAbs(laplacian_filtered)
plt.imshow(laplacian_filtered, cmap='gray')
plt.title('Laplacian Filtered Image')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/content/lowContrast.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
bilateral_filtered = cv2.bilateralFilter(image_rgb, d=9, sigmaColor=75, sigmaSpace=75)
plt.imshow(bilateral_filtered[...,::-1])
plt.title('Bilateral Filtered Image')
plt.show()
```
