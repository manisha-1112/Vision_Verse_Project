```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
def kmeans_segmentation(image, k_values):
    pixels = image.reshape((-1, 3))
    results = []
    for k in k_values:
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
        _, labels, centers = cv2.kmeans(pixels.astype(np.float32), k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
        centers = np.uint8(centers)
        segmented_image = centers[labels.flatten()]
        segmented_image = segmented_image.reshape(image.shape)
        results.append(segmented_image)
    return results
image_path = '/content/1.png'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
k_values = [3, 5, 8]
segmented_images = kmeans_segmentation(image, k_values)
plt.figure(figsize=(12, 4))
for i, k in enumerate(k_values):
    plt.subplot(1, 4, i + 2)
    plt.imshow(segmented_images[i])
    plt.title(f'Segmented (K={k})')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image_path = '/content/boat.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
_, otsu_threshold = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
_, normal_threshold = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(10, 7))
plt.subplot(1, 3, 2)
plt.imshow(otsu_threshold, cmap='gray')
plt.title('Otsu Thresholding')
plt.subplot(1, 3, 3)
plt.imshow(normal_threshold, cmap='gray')
plt.title('Normal Thresholding')
plt.show()
```
