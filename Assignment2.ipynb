```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def ds_image(image, factor):
    return cv2.resize(image, (image.shape[1] // factor, image.shape[0] // factor), interpolation=cv2.INTER_NEAREST)

image_path = r'/content/266689.jpg'
original_image = cv2.imread(image_path)
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
factors = [2, 4]

for i, factor in enumerate(factors):
    downsampled_image = ds_image(original_image, factor)

    plt.subplot(1, 3, i + 2)
    plt.title(f'Downsampled {factor}x')
    plt.imshow(cv2.cvtColor(downsampled_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def quantize_image(image, levels):


    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    quantization_step = 256 // levels
    quantized_image = (image // quantization_step) * quantization_step

    return quantized_image


image_path = r'/content/266689.jpg'
original_image = cv2.imread(image_path)

plt.figure(figsize=(10, 5))
plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

quantization_levels = [4, 8, 16]

for i, levels in enumerate(quantization_levels):
    quantized_image = quantize_image(original_image, levels)


    plt.subplot(1, 4, i + 2)
    plt.title(f'Quantized ({levels} levels)')
    plt.imshow(quantized_image, cmap='gray')
    plt.axis('off')

plt.show()
```
