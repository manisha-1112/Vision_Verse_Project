8. Print the first and the last element of the first row using stepsize approach.

```python
import numpy as np

arr = np.array([[1, 2, 3], [1, 4, 5], [8, 4, 2]])

first_and_last = arr[0, ::len(arr[0])-1]

print(first_and_last)
```

9. Print the maximum and minimum values of a.

```python
a = [[1, 2, 3], [1, 4, 5], [8, 4, 2]]

flatten_a = [val for row in a for val in row]
max_value = max(flatten_a)
min_value = min(flatten_a)

print( max_value)
print( min_value)
```

10. Print row-wise maximum values and their indicies of a.

```python
a = [[1, 2, 3], [1, 4, 5], [8, 4, 2]]

for i, row in enumerate(a):
    max_value = max(row)
    max_index = row.index(max_value)
    print(max_value)
    print(max_index)
```
