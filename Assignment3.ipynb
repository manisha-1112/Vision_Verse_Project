```python
import cv2
import numpy as np
image = cv2.imread(r"/content/2.jpeg")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

edges = cv2.Canny(gray, 50, 150)

lines = cv2.HoughLines(edges, 1, np.pi / 180, threshold=100)

for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image = cv2.imread('/content/2.jpeg', cv2.IMREAD_GRAYSCALE)

edges_sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
edges_sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

edges_prewitt_x = cv2.filter2D(image, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
edges_prewitt_y = cv2.filter2D(image, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

edges_robert_x = cv2.filter2D(image, -1, np.array([[1, 0], [0, -1]]))
edges_robert_y = cv2.filter2D(image, -1, np.array([[0, 1], [-1, 0]]))

cv2_imshow( edges_sobel_x)
cv2_imshow( edges_sobel_y)
cv2_imshow( edges_prewitt_x)
cv2_imshow( edges_prewitt_y)
cv2_imshow( edges_robert_x)
cv2_imshow( edges_robert_y)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
